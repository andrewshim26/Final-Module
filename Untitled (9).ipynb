```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Load datasets
mobile_usage = pd.read_csv('mobile_usage_behavioral_analysis.csv')
survey = pd.read_csv('survey.csv')
mental_health = pd.read_csv('mental_health_screen_time_dataset.csv')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select features for clustering
features = mental_health[['Daily_Screen_Time', 'App_Social_Media_Time', 'Sleep_Duration']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform k-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
mental_health['Cluster'] = kmeans.fit_predict(scaled_features)

# Cluster visualization
sns.pairplot(mental_health, hue='Cluster', vars=['Daily_Screen_Time', 'Well_Being_Score', 'Stress_Level'])
plt.title("Clustering of Screen Time and Mental Health Metrics")
plt.show()
```

```python
# Regression to predict Well_Being_Score
X = mental_health[['Daily_Screen_Time', 'Sleep_Duration', 'Phone_Unlocks']]
y = mental_health['Well_Being_Score']
X = sm.add_constant(X)  # Add constant for intercept
model = sm.OLS(y, X).fit()

# Display regression summary
print(model.summary())
```

```python
#1. Distribution of Daily Screen Time
sns.histplot(mobile_usage['Daily_Screen_Time_Hours'], kde=True, bins=30)
plt.title("Distribution of Daily Screen Time")
plt.xlabel("Daily Screen Time (Hours)")
plt.ylabel("Frequency")
plt.show()

#2. Scatterplot: Screen Time vs Well-Being
sns.scatterplot(data=mental_health, x='Daily_Screen_Time', y='Well_Being_Score', hue='Stress_Level')
plt.title("Screen Time vs Well-Being Score")
plt.xlabel("Daily Screen Time")
plt.ylabel("Well-Being Score")
plt.show()

# Filter for numerical columns only
numerical_columns = mental_health.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr_matrix = numerical_columns.corr()

#3. Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap: Mental Health Dataset")
plt.show()

#4. Countplot: Work Interference and Treatment
sns.countplot(data=survey, x='work_interfere', hue='treatment')
plt.title("Mental Health Treatment by Work Interference Levels")
plt.xlabel("Work Interference Level")
plt.ylabel("Count")
plt.show()
```
